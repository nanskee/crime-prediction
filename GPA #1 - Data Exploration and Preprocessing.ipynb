### **2. Data Cleaning**

#### 1. Handling Missing Values

```python
# Mocodes column (14.712434%)
# Filling missing `Mocodes` with "Unknown" prevents the loss of valuable 
# data from other columns (like crime type, location, or victim characteristics), 
# ensuring no critical information is discarded and maintaining the dataset's completeness.

crime['Mocodes'].fillna('Unknown', inplace=True)

# Verify if missing values are filled
print(crime['Mocodes'].isnull().sum())  # Should return 0 if all missing values are filled

```

```python
# Vict Sex column (14.02% missing)
# Since Vict Sex likely has a limited number of categories 
# (e.g., "M" or "F"), filling missing values with the most 
# frequent value (mode) could be a practical choice, preserving as 
# much data as possible.

# Get the mode of the 'Vict Sex' column
mode_value = crime['Vict Sex'].mode()[0]

# Fill the missing values with the mode
crime['Vict Sex'].fillna(mode_value, inplace=True)

# Verify if missing values are filled
print(crime['Vict Sex'].isnull().sum())  # Should return 0 if all missing values are filled
```

```python
# Vict Descent column (14.020547% missing)
# Since Vict Descent is a categorical variable
# (e.g., "Black" or "Chinese"), filling missing values with the most 
# frequent value (mode) could be a practical choice, preserving as 
# much data as possible.

# Get the mode of the 'Vict Descent' column
mode_value = crime['Vict Descent'].mode()[0]

# Fill the missing values with the mode
crime['Vict Descent'].fillna(mode_value, inplace=True)

# Verify if missing values are filled
print(crime['Vict Descent'].isnull().sum())  # Should return 0 if all missing values are filled
```

```python
# Weapon Used Cd column (66.692247% missing)
# Weapon Desc column (66.692247% missing)

# Fill missing values in Weapon Used Cd and Weapon Desc with 'Unknown'
crime['Weapon Used Cd'].fillna("Unknown", inplace=True)
crime['Weapon Desc'].fillna("Unknown", inplace=True)

# Verify if missing values are filled
print(crime['Weapon Used Cd'].isnull().sum())  # Should return 0 if all missing values are filled
print(crime['Weapon Desc'].isnull().sum())  # Should return 0 if all missing values are filled
```

```python
# Crm Cd 2 column (92.968115% missing)
# Crm Cd 3 column (99.764057% missing)
# Crm Cd 4 column (99.993460% missing)

# Since Crm Cd 2, Crm Cd 3, and Crm Cd 4 represent secondary, tertiary, 
# and quaternary crimes, these columns are supplementary and not essential 
# to the core crime analysis. Dropping them would not affect the analysis 
# of the primary crime (Crm Cd 1) or other key features (like victim data, 
# location, or weapon used).

# Dropping columns with such high levels of missing data is a reasonable 
# approach since keeping them adds very little information to the analysis 
# and unnecessarily increases dataset size.

# Dropping the columns 'Crm Cd 2', 'Crm Cd 3', and 'Crm Cd 4'
crime = crime.drop(['Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4'], axis=1)

# Verify that the columns have been dropped
print(crime.columns)

```

```python
# Cross Street column (84.526603% missing)

# With over 84% missing values, keeping the column may not add 
# significant value. If location analysis relies more on latitude, 
# longitude, or area names, this column could be dropped to reduce 
# dataset complexity and size without losing critical information.

# Dropping the 
crime = crime.drop(['Cross Street'], axis=1)

# Verify that the columns have been dropped
print(crime.columns)
```

### **3. Data Transformation**

#### 1. Date and Time Transformation

```python
# Transform the time and date columns (e.g., Date Occ and Time Occ) into datetime format 

# 1. Clean the 'DATE OCC' column to remove the time part
crime['DATE OCC'] = crime['DATE OCC'].str.split(' ').str[0]  # Keep only the date part

# 2. Combine cleaned 'DATE OCC' and 'TIME OCC' into a single datetime column
# Pad TIME OCC with leading zeros to ensure proper formatting (e.g., 2130 to 21:30)
crime['Datetime'] = pd.to_datetime(crime['DATE OCC'] + ' ' + crime['TIME OCC'].astype(str).str.zfill(4), format='%m/%d/%Y %H%M')

# Display the updated DataFrame with the new Datetime column
print(crime[['DATE OCC', 'TIME OCC', 'Datetime']])
```

```python
# Extract key information from the 'Datetime' column
crime['Day of Week'] = crime['Datetime'].dt.day_name()  # Extract day of the week
crime['Month'] = crime['Datetime'].dt.month  # Extract month (as integer)
crime['Hour'] = crime['Datetime'].dt.hour  # Extract hour (as integer)
crime['Year'] = crime['Datetime'].dt.year  # Extract year

# Display the updated DataFrame with the new extracted columns
print(crime[['Datetime', 'Day of Week', 'Month', 'Hour', 'Year']])
```

```python
import pandas as pd

# Sample DataFrame creation (you can skip this if you already have your DataFrame)
# crime = pd.DataFrame({'Hour': [5, 12, 19, 22, 14, 23, 6]})

# Define a function to classify time into 'Day' or 'Night'
def time_of_day(hour):
    if 6 <= hour < 18:  # Daytime is defined as 6 AM to 6 PM
        return 'Day'
    else:
        return 'Night'

# Apply the function to the 'Hour' column to create a new binary feature
crime['Time_of_Day'] = crime['Hour'].apply(time_of_day)

# Display the updated DataFrame
print(crime[['Hour', 'Time_of_Day']])
```
